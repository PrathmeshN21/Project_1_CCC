```python
# ✅ CCC Mandatory Fields Checker (Google Colab Version)

import pandas as pd

class CCCValidator:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)  # CSV for Colab
        self.errors = []

    def validate(self):
        for index, row in self.df.iterrows():
            row_num = index + 2

            # Rule 1: Product = P → Model Year = 2075
            if str(row.get("Product")).strip() == "P":
                if str(row.get("Model Year")).strip() != "2075":
                    self.errors.append(f"Row {row_num}: Model Year must be 2075 for Product P")

            # Rule 2: ANA Description → Mass = 0
            ana_desc = str(row.get("ANA Description", "")).lower()
            if "chart current mass" in ana_desc or "new mass" in ana_desc:
                current_mass = row.get("Current Mass", 0)
                new_mass = row.get("New Mass", 0)

                if current_mass != 0 or new_mass != 0:
                    self.errors.append(f"Row {row_num}: Mass must be 0 when ANA Description has mass keywords")

            # Rule 3: Part Type = D
            if str(row.get("Part Type")).strip() != "D":
                self.errors.append(f"Row {row_num}: Part Type must be D")

            # Rule 4: Part DRD = SD
            if str(row.get("Part DRD")).strip() != "SD":
                self.errors.append(f"Row {row_num}: Part DRD must be SD")

            # Rule 5: Engineer Code = PAN1
            if str(row.get("Engr Code")).strip() != "PAN1":
                self.errors.append(f"Row {row_num}: Engineer Code must be PAN1")

            # Rule 6: Model Code = PPPA
            if str(row.get("Model code")).strip() != "PPPA":
                self.errors.append(f"Row {row_num}: Model Code must be PPPA")

        return self.errors


# ✅ Step 1: Upload file manually
from google.colab import files
uploaded = files.upload()

# Get uploaded file name
file_path = list(uploaded.keys())[0]

# ✅ Step 2: Run validation
validator = CCCValidator(file_path)
errors = validator.validate()

# ✅ Step 3: Output
if errors:
    print("\n❌ Validation Errors Found:\n")
    for err in errors:
        print(err)
else:
    print("\n✅ All CCC Rules Satisfied!")
```
